```python
import pandas as pd
```

```python
def getDf(filepath):
    df = pd.read_csv(filepath)
    return df

df = getDf('data.csv')
```

```python
def getStopsDict(df):
    stops_data = {}

    for _, row in df.iterrows():
        start_stop_key = row["start_stop"]
        start_stops = [row["start_stop_lat"], row["start_stop_lon"]]

        if (start_stop_key not in stops_data):
            stops_data[start_stop_key] = [start_stops]
        else:
            if ( start_stops not in stops_data[start_stop_key] ):
                stops_data[start_stop_key].append(start_stops)

        end_stop_key = row["end_stop"]
        end_stops = [row["end_stop_lat"], row["end_stop_lon"]]

        if (end_stop_key not in stops_data):
            stops_data[end_stop_key] = [end_stops]
        else:
            if ( end_stops not in stops_data[end_stop_key] ):
                stops_data[end_stop_key].append(end_stops)

    return stops_data

stops_data = getStopsDict(df)
```

```python
average_coordinates = lambda coords: [sum(x)/len(coords) for x in zip(*coords)]

def unifyStops(stops_data: dict):
    for key, values in stops_data.items():
        stops_data[key] = average_coordinates(values)

    return stops_data

stops_data_unified = unifyStops(stops_data)
```

```python
def update_stop_coordinates(df, stops_data):
    for index, row in df.iterrows():
        if row['start_stop'] in stops_data:
            df.at[index, 'start_stop_lat'], df.at[index, 'start_stop_lon'] = stops_data[row['start_stop']]

        if row['end_stop'] in stops_data:
            df.at[index, 'end_stop_lat'], df.at[index, 'end_stop_lon'] = stops_data[row['end_stop']]

    return df

updated_df = update_stop_coordinates(df, stops_data_unified)
```
